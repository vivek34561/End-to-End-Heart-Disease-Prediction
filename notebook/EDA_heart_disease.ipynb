```python
import numpy as np
import pandas as pd

# Set random seed for reproducibility
np.random.seed(42)

# Example: simulate your dataset (use your actual DataFrame instead)
# df = pd.read_csv('your_dataset.csv')

# Define categories
categories = ['High', 'Medium', 'Low']

# Identify missing values in 'Alcohol Consumption'
mask = df['Alcohol Consumption'].isna()

# Randomly assign values with equal probability
df.loc[mask, 'Alcohol Consumption'] = np.random.choice(categories, size=mask.sum(), replace=True)
```

```python
# Handling missing values 
# gender -> mode
# exercise habits -> mode
# smoking -> mode
# Family Heart Disease -> mode
# diabeties -> mode
# High Blood Pressure -> mode
# Low HDL Cholesterol -> mode
# stress Level -> mode
# sugar Consumption -> mode
# fasting Blood Sugar -> mean
# Drop columns with too many missing values
threshold = 0.4  # 40%
missing_percent = df.isnull().sum() / len(df)
df = df.drop(columns=missing_percent[missing_percent > threshold].index)

# Fill categorical with mode
cat_cols = df.select_dtypes(include=['object', 'category']).columns
for col in cat_cols:
    df[col] = df[col].fillna(df[col].mode()[0])

# Fill numerical with median
num_cols = df.select_dtypes(include=['number']).columns
for col in num_cols:
    df[col] = df[col].fillna(df[col].median())


```
